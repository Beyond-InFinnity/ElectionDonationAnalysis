```python
info_states['living_costs'] = info_states['living_costs'].replace('[\$,]', '', regex=True).astype(int)
info_states['disposable_income'] = info_states['disposable_income'].replace('[\$,]', '', regex=True).astype(int)
```

```python
info_states['average_salary'] = info_states['living_costs'] + info_states['disposable_income']
info_states['individual_contributions'] = individual_contributions_df['individual_contributions']
```

```python
info_states.head()
```

```python
info_states.to_csv('info_states.csv', index=False)
```

```python
info_states_sorted = info_states.sort_values(by='average_salary', ascending=True)
num_states = len(info_states)
blue_gradient = plt.cm.Blues(np.linspace(0.3, 1, num_states))
color_mapping = {state: color for state, color in zip(info_states_sorted['State'], blue_gradient)}

plt.figure(figsize=(12, 15))
plt.barh(info_states['State'], info_states['average_salary'], color=[color_mapping[state] for state in info_states['State']])
plt.xlabel('Average Salary')
plt.ylabel('State')
plt.title('Average Salary by State')
plt.savefig('average_salary_by_state.png')
plt.show();
```

```python
info_states_sorted = info_states.sort_values(by='living_costs', ascending=True)
num_states = len(info_states)
blue_gradient = plt.cm.Blues(np.linspace(0.3, 1, num_states))
color_mapping = {state: color for state, color in zip(info_states_sorted['State'], blue_gradient)}

plt.figure(figsize=(12, 15))
plt.barh(info_states['State'], info_states['living_costs'], color=[color_mapping[state] for state in info_states['State']])
plt.xlabel('Cost of Living')
plt.ylabel('State')
plt.title('Cost of Living by State')
plt.savefig('cost_of_living_by_state.png')
plt.show();
```

```python
info_states_sorted = info_states.sort_values(by='disposable_income', ascending=True)
num_states = len(info_states)
blue_gradient = plt.cm.Blues(np.linspace(0.3, 1, num_states))
color_mapping = {state: color for state, color in zip(info_states_sorted['State'], blue_gradient)}

plt.figure(figsize=(12, 15))
plt.barh(info_states['State'], info_states['disposable_income'], color=[color_mapping[state] for state in info_states['State']])
plt.xlabel('Disposable Income')
plt.ylabel('State')
plt.title('Disposable Income by State')
plt.savefig('disposable_income_by_state.png')
plt.show();
```
